##merge air quality,greenspace



```python
import pandas as  pd
df1=pd.read_csv("../Data/raw/air_quality_2023.csv")
df2=pd.read_csv("../Data/raw/green_space.csv")
```

```python
airquality_greenspace = pd.merge(df1, df2, on='City')
```

```python
airquality_greenspace.to_csv("../Data/integrated/airquality_greenspace.csv",index=False)
```

##merge airquality_greenspace and Weather all cities

```python
df_airquality_greenspace=pd.read_csv("../Data/integrated/airquality_greenspace.csv")
df_weather_all_cities_2023=pd.read_csv("../Data/integrated/weather_all_cities_2023.csv")
df_electricity=pd.read_csv("../Data/raw/electricity_consumption_2023.csv")
```

```python
df_electricity.columns
```

```python
df_temp = pd.merge(df_airquality_greenspace, df_weather_all_cities_2023, on=['City', 'Date'])
df_all  = pd.merge(df_temp, df_electricity, on=['City', 'Date'])
```

```python
df_all.to_csv("../Data/integrated/all_data.csv",index=False)
```
